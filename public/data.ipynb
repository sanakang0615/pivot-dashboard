```python

import pandas as pd

# Load the AdPack data
adpack_df = pd.read_parquet('./db/adpack_data.parquet')
# Load the Campaign data
campaign_df = pd.read_parquet('./db/campaign_data.parquet')
```

```python
# Check if (account_id, campaign) pair is unique in adpack_df
is_adpack_pair_unique = adpack_df.set_index(['account_id', 'campaign']).index.is_unique
print(f"Is (account_id, campaign) pair unique in adpack_df? {is_adpack_pair_unique}")

# Check if (account_id, campaign) pair is unique in campaign_df
is_campaign_pair_unique = campaign_df.set_index(['account_id', 'campaign']).index.is_unique
print(f"Is (account_id, campaign) pair unique in campaign_df? {is_campaign_pair_unique}")
```
